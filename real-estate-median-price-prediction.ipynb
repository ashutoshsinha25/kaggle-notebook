```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestRegressor
```

```python
data = pd.read_csv('../input/real-estate-dataset/data.csv')
```

```python
data
```

```python
data.info()
```

```python
# preprocessing 
```

```python
data.isnull().sum()
```

```python
data['RM'] = data['RM'].fillna(data['RM'].mean())
```

```python
data.isna().sum().sum()
```

```python
y = data['MEDV']
X = data.drop('MEDV', axis=1)
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python
#traning 
```

```python
model = RandomForestRegressor(n_estimators=100, criterion='mse', bootstrap=False)
model.fit(X_train, y_train)
```

```python
bootstrap_model = RandomForestRegressor(n_estimators=100, criterion='mse', bootstrap=True)
bootstrap_model.fit(X_train, y_train)
```

```python
print("R^2 without bootstrapping:", model.score(X_test, y_test))
print("R^2 with bootstrapping:", bootstrap_model.score(X_test, y_test))
```
