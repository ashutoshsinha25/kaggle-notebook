```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
data = pd.read_csv('../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
```

```python
data
```

```python
profile = ProfileReport(data)
```

```python
profile.to_notebook_iframe()
```

```python
# spliting data
y = data['DEATH_EVENT']
X = data.drop('DEATH_EVENT' , axis = 1)
```

```python
y
```

```python
X 
```

```python
# scaling
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
pd.DataFrame(X)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y ,train_size = 0.7)
```

```python
lr = LogisticRegression()
svm = SVC()
nn = MLPClassifier(hidden_layer_sizes = (128 , 128))
```

```python
lr.fit(X_train , y_train)
lr_pred = lr.predict(X_test)

lr_acc = lr.score(X_test,y_test)
lr_f1 = f1_score(y_test , lr_pred)
```

```python
svm.fit(X_train , y_train)
svm_pred = svm.predict(X_test)

svm_acc = svm.score(X_test , y_test)
svm_f1 = f1_score(y_test , svm_pred)
```

```python
nn.fit(X_train , y_train)
nn_pred = nn.predict(X_test)

nn_acc = nn.score(X_test , y_test)
nn_f1 = f1_score(y_test , nn_pred)
```

```python
print(f"Logistic Regression: \nAccuracy : {lr_acc}\nF1 Score : {lr_f1}\n")
print(f"Support Vectore Machine: \nAccuracy : {svm_acc}\nF1 Score : {svm_f1}\n")
print(f"Neural Network: \nAccuracy : {nn_acc}\nF1 Score : {nn_f1}\n")
```
