### Problem Statement: Given data about daily sunrise and sunset, predict the month of a given record 

```python
import numpy as np
import pandas as pd 


from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression 


import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('/kaggle/input/sundown-and-sunset-data-20102020/Sunup_Sundown_Data.csv')
```

```python
data.info()
```

```python
data.head(4)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['month'] = data['Date'].apply(lambda x : x.month)
```

```python
data['month'].value_counts(normalize = True)
```

```python
data['SunupHr'] = data['Sunup'].apply(lambda x : x.split(':')[0])
data['SunupMin'] = data['Sunup'].apply(lambda x : x.split(':')[1])

data['SundownHr'] = data['Sundown'].apply(lambda x : x.split(':')[0])
data['SundownMin'] = data['Sundown'].apply(lambda x : x.split(':')[1])
```

```python
data.drop(['Date' , 'Sunup' , 'Sundown','City'],inplace=True,axis=1)
```

```python
data = data.astype(int)
```

```python
corr = data.corr()

plt.figure(figsize = (5,5)) 

sns.heatmap(corr , annot = True  ,vmin =-1.0 ,vmax = 1.0 , cmap='rocket')

plt.show()
```

```python
plt.figure(figsize = (10,5))
for i in data.columns : 
    sns.kdeplot(data[i] , fill=True , legend= True)
plt.show()
```

```python
X = data.drop('month',axis=1)
y = data['month']
```

```python
sc = StandardScaler()
```

```python
X = sc.fit_transform(X)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , train_size = 0.7 , random_state = 123)
```

```python
C = [0.01 , 0.1 , 1.0 , 10.0 ,100.0]

models = []

for i in C:
    lr = LogisticRegression(C=i)
    lr.fit(X_train , y_train)
    models.append(lr)
```

```python
# checking the accuracy of the models
[model.score(X_test , y_test) for model in models]
```
