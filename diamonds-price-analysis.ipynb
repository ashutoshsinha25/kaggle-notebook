```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
 
    
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler 

from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression, Ridge, Lasso 
```

```python
df = pd.read_csv('/kaggle/input/diamonds/diamonds.csv')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0',axis =1,inplace=True)
```

```python
df.head()
```

```python
y = df['price']
X = df.drop('price',axis=1)
```

```python
print(f'Cuts : {len(X["cut"].unique())}')
print(f'Colors : {len(X["color"].unique())}')
print(f'Clarity : {len(X["clarity"].unique())}')
```

```python
encoder = LabelEncoder()
X['cut'] = encoder.fit_transform(X['cut'])
cut_mapping = {index:label for index, label in enumerate(encoder.classes_)}
```

```python
cut_mapping
```

```python
X['color'] = encoder.fit_transform(X['color'])
color_mapping = {idx : lb for idx,lb in enumerate(encoder.classes_)}
X['clarity'] = encoder.fit_transform(X['clarity'])
clarity_mapping = {idx : lb for idx,lb in enumerate(encoder.classes_)}
```

```python
color_mapping
```

```python
clarity_mapping 
```

```python
X
```

```python
scaler = MinMaxScaler() 

X = scaler.fit_transform(X)
```

```python
pd.DataFrame(X)
```

```python
X_train,X_test , y_train,y_test = train_test_split(X , y , train_size = 0.8)
```

```python
model_lr = LinearRegression()
model_l1 = Lasso(alpha = 1)
model_l2 = Ridge(alpha = 1)
```

```python
model_lr.fit(X_train , y_train)
model_l1.fit(X_train , y_train)
model_l2.fit(X_train , y_train)
```

```python
print(f'Linear Model without regularization score : {model_lr.score(X_test , y_test)}')
print(f'Linear Model with L1 regularization score : {model_l1.score(X_test , y_test)}')
print(f'Linear Model with L2 regularization score : {model_l2.score(X_test , y_test)}')
```

```python
for i in list(np.round(np.linspace(0,2,50),1)):
    model_l2 = Ridge(alpha = i)
    model_l2.fit(X_train , y_train)
    print(f'Linear Model with L2 regularization score (exp) : {model_l2.score(X_test , y_test)}')
```
